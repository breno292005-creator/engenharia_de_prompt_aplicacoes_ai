### Soma de Lista

```python
numeros = [1, 2, 3, 4, 5]
soma = 0
i = 0
while i < len(numeros):
  soma += numeros[i]
  i += 1
print(numeros)
print(soma)
```

```python
numeros = [1, 2, 6, 9, 11] # Inicializa uma lista de números inteiros
soma = 0 # Inicializa uma variável para armazenar a soma

# Itera sobre cada número na lista 'numeros'
for num in numeros:
  soma += num # Adiciona o número atual à soma

print(soma) # Imprime o valor final da soma
```

### Eh primo

```python
def eh_primo(n):
    if n <= 1:
        return False

    for i in range(2, n):
        if n % i == 0:
            return False

    return True

# Teste
numero = int(input("Digite um número: "))

if eh_primo(numero):
    print("É primo")
else:
    print("Não é primo")
```

### Inversão String

```python
def inverte_texto(texto):
  invertido = ''
  for letra in texto:
    invertido = letra + invertido
  return invertido
```

```python
# Calculadora
a1 = float(input("Digite um valor: "))
b = input("Digite sinal (+) ou (-): ")
a2 = float(input("Digite segundo valor: "))
if b == "+":
  resultado = a1 + a2
  print("Resultado: ", resultado)
elif b == "-":
  resultado = a1 - a2
  print("Resultado: ", resultado)
else:
  print("Operação invalida, coloque como sinal: (+) ou (-)")
```

```python
num1 = float(input("Digite o primeiro número: "))
operacao = input("Digite a operação (+), (-), (*) ou (/): ")
num2 = float(input("Digite o segundo número: "))

if operacao == "somar":
    resultado = num1 + num2
    print("Resultado:", resultado)

elif operacao == "subtrair":
    resultado = num1 - num2
    print("Resultado:", resultado)

elif operacao == "multiplicação":
    resultado = num1 * num2
    print("Resultado:", resultado)

elif operacao == "divisão":
    if num2 != 0:
        resultado = num1 / num2
        print("Resultado:", resultado)
    else:
        print("Erro: divisão por zero não é permitida")

else:
    print("Operação inválida")
```
